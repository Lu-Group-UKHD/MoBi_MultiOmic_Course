# Setting up Python with `conda`

For processing and analyzing single-cell data, we will be using Python as part of this lecture. For reproducible Python workflows we use `conda`, which loads and installs a mini-system and Python inside. This includes software, which would otherwise be installed on your computer (e.g. `hdf5`)

`conda` should not be confused with "Anaconda", the company. `conda` is an open-source software, which installs pre-compiled binaries (what compilation produces), Anaconda is a company hosting thees binaries on servers.

Because Anaconda is commercial, alternative channels have emerged, such as `conda-forge` and `bioconda`, which we will be using.

Please note, `bioconda` hosts Python and R-packages we require, but is not available for Windows. Therefore, seriously consider abandoning Windows entirely, otherwise dual-boot will do (https://www.youtube.com/watch?v=qq-7X8zLP7g). Windows-only Users can´t be helped.

The exact software we want to download is `miniforge3`. This is basically the same as `conda`, or rather `miniconda` (fewer pre-installed packages), but natively includes both the `conda` and `mamba` commands. If you ever see something done using the `conda` command, you can replace that with `mamba` and it will be faster.

For Unix-based systems, the most robust method is using the shell script installer. These commands automatically detect your architecture (Intel vs. Apple Silicon/ARM). In a terminal, run:

```python
#| eval: false

# 1. Download the latest installer script
curl -L -O "https://github.com/conda-forge/miniforge/releases/latest/download/Miniforge3-$(uname)-$(uname -m).sh"

# 2. Run the installer (Interactive)
bash Miniforge3-$(uname)-$(uname -m).sh

# OR: Run a "Silent" install (Batch mode) to a specific directory
# bash Miniforge3-$(uname)-$(uname -m).sh -b -p ~/miniforge3
```

Initialize the Shell

After installation, you must initialize your shell so the conda command is recognized:

```python
#| eval: false

~/miniforge3/bin/conda init  # Use the path where you installed it
source ~/.bashrc             # For Linux (or ~/.zshrc for macOS)
```

For Windows users, these are the commands to install. However, I cannot verify this due to lack of Windows computers I can test on. Remember, that `bioconda` packages cannot be used on Windows.

```python
#| eval: false
# 1. Download the installer via PowerShell
Invoke-WebRequest -Uri "https://github.com/conda-forge/miniforge/releases/latest/download/Miniforge3-Windows-x86_64.exe" -OutFile "Miniforge3.exe"

# 2. Run silent installation
Start-Process -FilePath ".\Miniforge3.exe" -ArgumentList "/S", "/InstallationType=JustMe", "/D=$HOME\miniforge3" -Wait

# Initialize
& "$HOME\miniforge3\condabin\conda.bat" init
```

Now you can install `conda`-environments with Python:

```python
#| eval: false
conda create -n test_env python=3.12

# or we do
mamba create -n test_env python=3.12
```

This command creates the environment `test_env`, with python 3.12.x installed inside. There is always an environment active. Without expilitely activating and env, "base" is active, your own computer/system. We want to activate the environment, and then start installing into the environment:

```python
#| eval: false
conda activate test_env
conda install conda-forge::scanpy conda-forge::pandas conda-forge::numpy conda-forge::igraph
mamba install decoupler-py

pip install biocutils
pip install singlecellexperiment
pip install scranpy
```

`conda` and `mamba` install pre-compiled from one of the channels, `pip` builds the packages from the ground up (like running Python-files when installing). 

Note: earlier I mentioned installing from Anaconda, conda-forge, or bioconda. With miniforge, we automatically install from conda-forge, so you don't need to worry.

For reproducibitly, we want to write and use `.yaml` files. With the command below, we install an environment and install every package, including versions from a file we specify:

```python
#| eval: false
conda env create --name env_from_yaml -f /path/to/file.yaml
```

Importantly, how do we conveniently write code? I suggest JupyterLab. Jupyter Notebooks are a long-standing standard for itneractive use of Python. They are similar to RMarkdown, but have their own gimmicks. We can use them for running Python, or R if we wish.

JupyterLab is the package we install to get access to Jupyter Notebooks:

```python
#| eval: false
conda install conda-forge::jupyterlab conda-forge::ipywidgets
```

When installed we first need to start Jupyter. But instead of an own window, it uses the browser to display everything we need:

```python
#| eval: false
jupyter lab
```

When we are done with the session:

```python
#| eval: false
conda deactivate
```

And when we want to remove the environment:

```python
#| eval: false
conda remove --name test_env --all
```
