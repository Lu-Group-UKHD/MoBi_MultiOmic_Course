## 4.2 RNA preprocessing

Complete or inspect the preprocessing workflow below.

```python
# Store raw counts if they are not already stored
if "counts" not in rna.layers:
    rna.layers["counts"] = rna.X.copy()

# Normalize and log-transform
sc.pp.normalize_total(rna, target_sum=1e4)
sc.pp.log1p(rna)

# Select highly variable genes
sc.pp.highly_variable_genes(
    rna,
    n_top_genes=3000,
    flavor="seurat",
)

# PCA on highly variable genes
sc.pp.pca(
    rna,
    n_comps=50,
    use_highly_variable=True,
)

# Neighborhood graph, UMAP, and clustering
sc.pp.neighbors(rna, n_neighbors=15, n_pcs=50)
sc.tl.umap(rna)
sc.tl.leiden(rna, resolution=0.8, key_added="leiden_rna_0.8")
```
